```python
import numpy as np
import pandas as pd

```

```python
bn = pd.read_csv('BankNote_Authentication.csv')
```

```python
bn
```

```python
X = bn.iloc[:, :-1]
Y = bn.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.7, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

bnClassifier = RandomForestClassifier();

bnClassifier.fit(X_train, Y_train);
```

```python
y_pred = bnClassifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(Y_test, y_pred)
print(score)
```

```python
import pickle
```

```python
pickle_out = open('bnClassifier.pkl', 'wb')
pickle.dump(bnClassifier, pickle_out)
pickle_out.close()
```
